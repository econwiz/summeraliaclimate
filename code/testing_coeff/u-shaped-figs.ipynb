```python
import numpy as np
import matplotlib.pyplot as plt

# Regression coefficients from Stata
b0 = 683.0932
b1 = 0.0089803
b2 = 0.0017162
b3 = -0.0002545
b4 = 5.83e-06

# Range of annual average temperatures to plot
T = np.linspace(-5, 35, 400)  # Assuming °C, adjust if using °F

# Predicted mortality rate (deathrate_w99)
y_hat = b0 + b1*T + b2*T**2 + b3*T**3 + b4*T**4

# Plot
plt.figure(figsize=(8, 5))
plt.plot(T, y_hat, label="Predicted Death Rate", color='crimson')
plt.axvline(x=T[np.argmin(y_hat)], color='gray', linestyle='--', label='Min Mortality')
plt.xlabel("Annual Average Temperature (°C)")
plt.ylabel("Predicted Annual Mortality (deaths per million)")
plt.title("Estimated Temperature–Mortality Relationship (ERA5)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# GMFD Regression Coefficients
b0 = 643.03
b1 = -0.0166796
b2 = 0.0017074
b3 = -0.0000831
b4 = 1.49e-06

T = np.linspace(-5, 35, 400)

y_hat = b0 + b1*T + b2*T**2 + b3*T**3 + b4*T**4

# Plot
plt.figure(figsize=(8, 5))
plt.plot(T, y_hat, label="GMFD Fitted Curve", color='navy')
plt.axvline(x=T[np.argmin(y_hat)], color='black', linestyle='--', label='Min Mortality')
plt.xlabel("Annual Average Temperature (°C)")
plt.ylabel("Predicted Annual Mortality (deaths per million)")
plt.title("Estimated Temperature–Mortality Relationship (GMFD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Spec 2 MERRA2 coefficients (Table 3, Col 2 from your screenshot)
b0 = 774.9235      # intercept (_cons)
b1 = -0.0234       # tavg_poly_1_MERRA2
b2 = -0.000380     # tavg_poly_2_MERRA2
b3 = 0.000000348   # tavg_poly_3_MERRA2
b4 = 0.000000539   # tavg_poly_4_MERRA2

# Temperature grid (°C)
T = np.linspace(-5, 35, 400)

# Compute fitted deathrate_w99
y_hat = b0 + b1*T + b2*T**2 + b3*T**3 + b4*T**4

# Plot
plt.figure(figsize=(8, 5))
plt.plot(T, y_hat, color='forestgreen', label='MERRA2 Spec 2')
min_T = T[np.argmin(y_hat)]
plt.axvline(min_T, linestyle='--', color='black',
            label=f'Min Mortality at {min_T:.1f}°C')
plt.xlabel("Annual Average Temperature (°C)")
plt.ylabel("Predicted Mortality (deathrate_w99, deaths per million)")
plt.title("Temperature–Mortality Curve (MERRA2, Spec 2)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
