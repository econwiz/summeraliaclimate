```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from cartopy import crs as ccrs

# Path to your existing .zarr file
zarr_path = "/user/ab5405/summeraliaclimate/data/ACCESS-ESM1-5/tasdmgfparams_day_ACCESS-ESM1-5_ssp245_r10i1p1f1_ALLGWLS_projQDM-baseERA5-025_dwnsclQPLAD-target025deg_CONUS.zarr"
```

```python
ds = xr.open_zarr(zarr_path, consolidated=False)
ds
```

```python
import xarray as xr
import numpy as np

# Table 2 damage coefficients
dmgf_vals = [
     0.000234,  0.000126, -0.000144, -0.000269, -0.000322, -0.000195,
    -0.000119, -0.000074, -0.000003, -0.000036,  0.000000, -0.000111,
    -0.000311, -0.000294, -0.000585, -0.000646, -0.000757
]

# Assign to your bin_f values (must match length!)
dmgf_dnh = xr.DataArray(
    dmgf_vals,
    dims=["bin_f"],
    coords={"bin_f": ds['bin_f'].values[:17]}  # only use first 17 bins
)

# Compute impact at 2.5°C GWL
impact_25 = (ds['bins_behrer'].sel(gwl=2.5, bin_f=dmgf_dnh['bin_f']) * dmgf_dnh).sum(dim='bin_f') * 100
impact_25.plot()
plt.title("Projected % Income Impact at GWL 2.5°C")
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

# Set up map projection (PlateCarree is standard for lat/lon data)
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())

# Plot the data
pc = impact_25.plot.pcolormesh(
    ax=ax,
    transform=ccrs.PlateCarree(),  # tells cartopy your data is in lat/lon
    cmap='coolwarm',
    add_colorbar=True
)

# Add map features
ax.coastlines(resolution='10m', linewidth=1)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.set_title("% Impact on GDP at 2.5°C GWL (with Coastlines)")

plt.show()
```
