```python
import matplotlib.pyplot as plt

# 1) Your GWL levels
gwl = [1.0, 1.5, 2.0, 2.5, 3.0]

# 2) Mortality fractions
f_internal_mort = [0.575697, 0.440232, 0.208426, 0.161265, 0.107334]
f_model_mort    = [0.424303, 0.559768, 0.791574, 0.838735, 0.892666]

# 3) GDP fractions
f_internal_gdp  = [0.502662, 0.535638, 0.408623, 0.316119, 0.218661]
f_model_gdp     = [0.497338, 0.464362, 0.591377, 0.683881, 0.781339]

# 4) Plot Mortality
plt.figure()
plt.bar(gwl, f_model_mort,    width=0.4, label='model')
plt.bar(gwl, f_internal_mort, width=0.4, bottom=f_model_mort, label='internal')
plt.xlabel('Global‐warming level (°C)')
plt.ylabel('Fraction of variance')
plt.title('Mortality uncertainty fractions by GWL')
plt.xticks(gwl)
plt.legend()
plt.tight_layout()

# 5) Plot GDP
plt.figure()
plt.bar(gwl, f_model_gdp,    width=0.4, label='model')
plt.bar(gwl, f_internal_gdp, width=0.4, bottom=f_model_gdp, label='internal')
plt.xlabel('Global‐warming level (°C)')
plt.ylabel('Fraction of variance')
plt.title('Per‐capita GDP uncertainty fractions by GWL')
plt.xticks(gwl)
plt.legend()
plt.tight_layout()

plt.show()
```
