```python
import numpy as np
import matplotlib.pyplot as plt

b0 = 683.0932
b1 = 0.0089803
b2 = 0.0017162
b3 = -0.0002545
b4 = 5.83e-06

T = np.linspace(-5, 35, 400)

y_annual = b0 + b1*T + b2*T**2 + b3*T**3 + b4*T**4

y_min = np.min(y_annual)
delta_annual = y_annual - y_min

delta_daily = delta_annual / 365

fig, ax1 = plt.subplots(figsize=(9, 5))

color1 = 'tab:blue'
ax1.plot(T, delta_annual, color=color1, label='Annual Δ Mortality')
ax1.set_xlabel("Annual Average Temperature (°C)")
ax1.set_ylabel("Δ Mortality (deaths per 100k per year)", color=color1)
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.plot(T, delta_daily, color=color2, linestyle='--', label='Daily Δ Mortality')
ax2.set_ylabel("Δ Mortality (deaths per 100k per day)", color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

plt.title("Temperature-Related Change in Mortality (Spec 2)")
plt.grid(True)
fig.tight_layout()
plt.show()
```
