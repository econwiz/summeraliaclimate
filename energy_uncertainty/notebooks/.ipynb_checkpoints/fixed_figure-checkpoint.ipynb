```python
"""
plot_energy_response.py

Plots the energy–temperature response function from the Rode et al. (2021)
interacted FD specification, estimated across four observational climate
products (GMFD, ERA5, JRA-3Q, MERRA2).

Produces three figures:
  1. New data: 3×3 panel (income tercile × climate tercile) with all four
     products overlaid.
  2. Original data: same panel layout using the CIL release coefficients.
  3. Comparison: original (blue) vs new GMFD (red) side-by-side in each cell.

Each plot is saved as a PDF to OUTDIR.

Inputs
------
- Per-product coefficient CSVs:
    NEW_STER / FD_FGLS_inter_{MODEL_NAME}_{product}_coeff.csv
    (produced by 3_FD_interacted_all.do)
- Original CIL coefficient CSV:
    ORIG_STER / FD_FGLS_inter_{MODEL_NAME}_coeff.csv
- Break data (income knot + tercile CDD/HDD/income means):
    BREAK_DATA = energy_data_release_2021oct21/DATA/regression/break_data_TINV_clim.dta


"""

import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# ============================================================
# PATHS
# ============================================================
EU  = Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty")
EDR = Path("/user/ab5405/summeraliaclimate/code/energy_consumption/"
           "energy_data_release_2021oct21")

# Coefficient CSVs produced by 3_FD_interacted_all.do
NEW_STER = EU / "data" / "regression" / "sters"

# Coefficient CSVs from the original CIL Zenodo release
ORIG_STER = EU / "output_original" / "sters"

# Break data from the CIL release — contains income knot location and
# population-weighted tercile means for CDD, HDD, and log income.
BREAK_DATA = EDR / "DATA" / "regression" / "break_data_TINV_clim.dta"

OUTDIR = EU / "figures" / "comparison"
OUTDIR.mkdir(parents=True, exist_ok=True)

# ============================================================
# SETTINGS — edit these to change what gets plotted
# ============================================================

# Energy product to plot on the y-axis response
FUEL = "electricity"   # "electricity" or "other_energy"

# Year at which to evaluate the temporal trend interactions.
# The interacted spec includes temp×year and temp×year² terms to capture
# how the temperature response shifts over time (e.g. adaptation). Setting
# YEAR_PLOT = 2099 traces out the end-of-century response.
YEAR_PLOT = 2099

# Which model type to read coefficients for.
# "global"     — simple quartic polynomial, no climate/income/year interactions
#                (matches 2_FD_uninteracted_all.do output)
# "interacted" — full interacted spec with TINV, income spline, year trends
#                (matches 3_FD_interacted_all.do output)
MODEL_TYPE = "interacted"

# Model name string used in the .ster and CSV filenames.
# Must match the `local model_name` set in the Stata regression script.
MODEL_NAME = "TINV_clim_quadinter"

# Temperature grid for plotting (°C)
T = np.arange(-5, 36, 1.0)

# Indicators for above/below 20°C threshold.
# Used to apply the correct sign convention to the CDD/HDD heterogeneity terms:
#   above20 × b_cdd × subCDD × (T^k - 20^k)   [cooling demand, positive above 20]
#   below20 × b_hdd × subHDD × (20^k - T^k)   [heating demand, positive below 20]
ABOVE20 = (T >= 20).astype(float)
BELOW20 = (T  < 20).astype(float)

# ============================================================
# HELPERS
# ============================================================

def read_coeff_csv(path: Path) -> pd.Series:
    """
    Read a coefficient CSV (columns: parm, beta) exported from Stata.
    Returns a Series indexed by coefficient name with float values.
    Strips whitespace from names since Stata sometimes pads them.
    """
    df = pd.read_csv(path)
    df["parm"] = df["parm"].str.strip()
    return pd.Series(df["beta"].values, index=df["parm"].values)


def get_coeff(coef: pd.Series, name: str, default: float = 0.0) -> float:
    """
    Look up a single coefficient by name. Returns `default` (0.0) if not
    found — this is intentional: omitted/collinear terms are stored as 0
    in the CSV, and missing terms (e.g. a suffix mismatch) should also
    contribute 0 rather than raising an error. Always verify that the
    terms you expect are actually present using print(coef.index) if
    the response curve looks wrong.
    """
    if name in coef.index:
        return float(coef[name])
    return float(default)


def ibar_from_breakdata(bd: pd.DataFrame, fuel: str) -> float:
    """
    Extract the income spline knot location (ibar) from the CIL break_data.

    The CIL income spline has two segments separated at the maximum income
    of large income group 1 (the lower group). The knot is the same for all
    regions within a fuel type — it is set by the global income distribution
    at the time of estimation. The dc1 spline term is then:
        dc1 = lgdppc_MA15 - ibar
    which equals 0 at the knot and is negative below it (group 1 only).
    """
    maxcol = f"maxInc_largegpid_{fuel}"
    flagcol = f"largegpid_{fuel}"
    return float(bd.loc[bd[flagcol] == 1, maxcol].max())


def get_tercile_means(bd: pd.DataFrame) -> tuple[dict, dict, dict]:
    """
    Extract population-weighted tercile means from the CIL break_data.

    Returns three dicts keyed by tercile ID (integer, 1=lowest, 3=highest):
      cdd_means : mean long-run CDD20 for each climate tercile (tpid)
      hdd_means : mean long-run HDD20 for each climate tercile (tpid)
      inc_means : mean log GDP per capita for each income tercile (tgpid)

    These are used as the representative covariate values when tracing out
    the response function for each 3×3 cell in the panel plot.

    NOTE: tpid=1 is the coldest tercile, tpid=3 is the hottest. 
    """
    cdd_means = bd.groupby("tpid")["avgCDD_tpid"].mean().to_dict()
    hdd_means = bd.groupby("tpid")["avgHDD_tpid"].mean().to_dict()
    inc_means = bd.groupby("tgpid")["avgInc_tgpid"].mean().to_dict()
    return cdd_means, hdd_means, inc_means


# ============================================================
# RESPONSE FUNCTIONS
# ============================================================

def yhat_global(coef: pd.Series, fuel: str, suffix: str = "") -> np.ndarray:
    """
    Evaluate the GLOBAL (uninteracted) energy response curve.

    Specification: quartic polynomial in temperature, interacted with
    product group (pg) and flow type (indf1). No climate, income, or
    year interactions. Matches the ininteracted_all.do output.

    Parameters
    ----------
    coef   : coefficient Series from read_coeff_csv()
    fuel   : "electricity" (pg=1) or "other_energy" (pg=2)
    suffix : "" for new pipeline (unsuffixed FD vars), "_GMFD" for original
             CIL release (where climate product was embedded in var names)

    Returns
    -------
    y : array of shape (len(T),) giving ΔE at each temperature point
    """
    pg = 1 if fuel == "electricity" else 2
    y  = np.zeros_like(T, dtype=float)

    for k in [1, 2, 3, 4]:
        b_temp = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_temp{k}{suffix}")
        y += b_temp * (T ** k)

    return y


def yhat_interacted(
    coef:    pd.Series,
    fuel:    str,
    subInc:  float,
    subCDD:  float,
    subHDD:  float,
    ibar:    float,
    year:    int,
    suffix:  str = "",
    max_k:   int = 2,
) -> np.ndarray:
    """
    Evaluate the INTERACTED energy response curve for a given covariate cell.

    Specification: matches 3_FD_interacted_all.do (Rode et al. quadinter).
    For each polynomial order k, the response at temperature T is:

        y(T) = b_temp * (T^k - 20^k)
             + above20 * b_cdd * subCDD * (T^k - 20^k)     [cooling heterogeneity]
             + below20 * b_hdd * subHDD * (20^k - T^k)     [heating heterogeneity]
             + b_dc1 * deltacut * (T^k - 20^k)             [income spline]
             + b_yt  * (T^k - 20^k) * year                 [linear time trend]
             + b_dc1y * deltacut * year * (T^k - 20^k)     [income × year]
             + b_y2t  * (T^k - 20^k) * year²               [quadratic time trend]
             + b_dc1y2 * deltacut * year² * (T^k - 20^k)   [income × year²]

    All terms are normalized so the response passes through 0 at T=20°C,
    consistent with the Stata plotting script which uses (temp^k - 20^k).

    Parameters
    ----------
    coef   : coefficient Series
    fuel   : "electricity" or "other_energy"
    subInc : representative log income for this cell (from inc_means)
    subCDD : representative long-run CDD20 for this cell (from cdd_means)
    subHDD : representative long-run HDD20 for this cell (from hdd_means)
    ibar   : income spline knot location (from ibar_from_breakdata)
    year   : year at which to evaluate temporal trend terms (e.g. 2099)
    suffix : "" for new pipeline, "_GMFD" for original CIL release
    max_k  : highest polynomial order to include (2 matches Stata plotting
             script; the regression uses k=1..4 but only k=1..2 are plotted
             in the original Figure 1C)

    Returns
    -------
    y : array of shape (len(T),)

    NOTE: if FD_yeartemp terms are collinear with the absorbed year FE in
    Stata (as observed in the quadinter spec), those coefficients will be
    0 in the CSV and the year/year² terms will contribute nothing here.
    This is correct behavior — do not substitute nonzero values.
    """
    pg       = 1 if fuel == "electricity" else 2
    ig       = 2 if subInc > ibar else 1   # large income group (1=lower, 2=upper)
    deltacut = subInc - ibar               # spline value: 0 at knot, negative below

    y = np.zeros_like(T, dtype=float)

    for k in range(1, max_k + 1):
        Tk       = T ** k
        dT       = Tk - (20.0 ** k)        # normalized: 0 at 20°C
        dT_below = (20.0 ** k) - Tk        # sign-flipped for below-20 heating terms

        # Base temperature response
        b_temp = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_temp{k}{suffix}")

        # Long-run climate heterogeneity (TINV cross terms)
        b_cdd  = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_cdd20_TINVtemp{k}{suffix}")
        b_hdd  = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_hdd20_TINVtemp{k}{suffix}")

        # Income spline interaction — no suffix; FD vars are unsuffixed in new pipeline
        b_dc1  = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_dc1_lgdppc_MA15I{ig}temp{k}")

        # Linear year interaction (will be 0 if collinear with year FE)
        b_yt   = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_yeartemp{k}{suffix}")
        b_dc1y = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_dc1_lgdppc_MA15yearI{ig}temp{k}")

        # Quadratic year interaction (quadinter only)
        b_y2t   = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_year2temp{k}{suffix}")
        b_dc1y2 = get_coeff(coef, f"c.indp{pg}#c.indf1#c.FD_dc1_lgdppc_MA15year2I{ig}temp{k}")

        # Accumulate contribution of polynomial order k
        y += b_temp  * dT
        y += ABOVE20 * b_cdd  * subCDD   * dT
        y += BELOW20 * b_hdd  * subHDD   * dT_below
        y += b_dc1   * deltacut          * dT
        y += b_yt    * year              * dT
        y += b_dc1y  * deltacut * year   * dT
        y += b_y2t   * (year ** 2)       * dT
        y += b_dc1y2 * deltacut * (year ** 2) * dT

    return y


# ============================================================
# LOAD BREAK DATA
# ============================================================
print("Loading break_data...")
bd = pd.read_stata(BREAK_DATA)

ibar                          = ibar_from_breakdata(bd, FUEL)
cdd_means, hdd_means, inc_means = get_tercile_means(bd)

# Diagnostic: print these to verify ordering before trusting the plots.
# tpid=1 should be coldest (lowest CDD, highest HDD), tpid=3 hottest.
# tgpid=1 should be lowest income, tgpid=3 highest.
print(f"Income knot (ibar): {ibar:.4f}")
print(f"CDD means by tpid:  {cdd_means}")
print(f"HDD means by tpid:  {hdd_means}")
print(f"Inc means by tgpid: {inc_means}")

# ============================================================
# SELECT PLOTTING FUNCTION
# ============================================================
if MODEL_TYPE == "global":
    print("\nUsing GLOBAL specification (quartic polynomial, no interactions)")
    needs_climate_income = False

elif MODEL_TYPE == "interacted":
    print(f"\nUsing INTERACTED specification ({MODEL_NAME})")
    needs_climate_income = True

else:
    raise ValueError(f"Unknown MODEL_TYPE: {MODEL_TYPE!r}. Use 'global' or 'interacted'.")

# ============================================================
# PLOT 1 — NEW DATA (all four products)
# ============================================================
print("\n" + "=" * 60)
print("PLOTTING NEW DATA (all products)")
print("=" * 60)

products_new = ["GMFD", "ERA5", "JRA_3Q", "MERRA2"]
coefs_new    = {}

for prod in products_new:
    if MODEL_TYPE == "global":
        coef_path = NEW_STER / f"FD_FGLS_global_{prod}_coeff.csv"
    else:
        coef_path = NEW_STER / f"FD_FGLS_inter_{MODEL_NAME}_{prod}_coeff.csv"

    if coef_path.exists():
        coefs_new[prod] = read_coeff_csv(coef_path)
        print(f"  Loaded {prod}: {len(coefs_new[prod])} coefficients")
    else:
        print(f"  WARNING: {coef_path} not found — skipping {prod}")

if needs_climate_income:
    fig, axes = plt.subplots(3, 3, figsize=(12, 9), sharex=True, sharey=True)
    fig.suptitle(f"New data: {FUEL} | {MODEL_NAME} | Year={YEAR_PLOT}", y=0.98)

    # Row order: high income (lg=3) top, low income (lg=1) bottom
    # Col order: hot climate (tr=3) left, cold (tr=1) right
    # This matches the Stata plotting script: forval lg=3(-1)1, forval tr=3(-1)1
    for r, lg in enumerate([3, 2, 1]):
        for c, tr in enumerate([3, 2, 1]):
            ax     = axes[r, c]
            subInc = inc_means[lg]
            subCDD = cdd_means[tr]
            subHDD = hdd_means[tr]

            for prod in products_new:
                if prod not in coefs_new:
                    continue
                y = yhat_interacted(
                    coefs_new[prod], FUEL, subInc, subCDD, subHDD,
                    ibar, YEAR_PLOT, suffix=""
                )
                ax.plot(T, y, linewidth=2, label=prod, alpha=0.8)

            ax.axhline(0, color="black", linewidth=0.5, linestyle="--")
            ax.axvline(20, color="gray",  linewidth=0.5, linestyle=":")
            ax.set_title(f"Inc={subInc:.2f}, CDD={subCDD:.0f}", fontsize=9)
            ax.grid(True, alpha=0.3)

            if c == 0:
                ax.set_ylabel("ΔE", fontsize=10)
            if r == 2:
                ax.set_xlabel("Temperature (°C)", fontsize=10)

    handles, labels = axes[0, 0].get_legend_handles_labels()
    fig.legend(handles, labels, loc="center right", frameon=False)
    plt.tight_layout(rect=[0, 0, 0.92, 0.96])

else:
    # Global spec: single panel, all products on same axes
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_title(f"New data: {FUEL} | Global model", fontsize=14)

    for prod in products_new:
        if prod not in coefs_new:
            continue
        y = yhat_global(coefs_new[prod], FUEL, suffix="")
        ax.plot(T, y, linewidth=2, label=prod, alpha=0.8)

    ax.axhline(0, color="black", linewidth=0.5, linestyle="--")
    ax.axvline(20, color="gray",  linewidth=0.5, linestyle=":")
    ax.set_xlabel("Temperature (°C)", fontsize=12)
    ax.set_ylabel("ΔE", fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.legend(frameon=False)
    plt.tight_layout()

out1 = (OUTDIR / f"NEW_all_products_{FUEL}_global.pdf"
        if MODEL_TYPE == "global" else
        OUTDIR / f"NEW_all_products_{FUEL}_{MODEL_NAME}.pdf")
fig.savefig(out1)
print(f"\nSaved: {out1.name}")

# ============================================================
# PLOT 2 — ORIGINAL CIL DATA
# ============================================================
print("\n" + "=" * 60)
print("PLOTTING ORIGINAL DATA (CIL release)")
print("=" * 60)

orig_coef_path = (ORIG_STER / f"FD_FGLS_global_TINV_clim_coeff.csv"
                  if MODEL_TYPE == "global" else
                  ORIG_STER / f"FD_FGLS_inter_{MODEL_NAME}_coeff.csv")

coef_orig   = None
suffix_orig = ""

if orig_coef_path.exists():
    coef_orig = read_coeff_csv(orig_coef_path)
    print(f"  Loaded ORIGINAL: {len(coef_orig)} coefficients")

    # The original CIL pipeline embedded the product suffix in FD variable
    # names (e.g. FD_temp1_GMFD). Detect this automatically so the response
    # function lookup uses the correct coefficient name format.
    has_suffix  = any("_GMFD" in name for name in coef_orig.index)
    suffix_orig = "_GMFD" if has_suffix else ""
    print(f"  Coefficient suffix: {suffix_orig!r if suffix_orig else '(none)'}")

    if needs_climate_income:
        fig2, axes2 = plt.subplots(3, 3, figsize=(12, 9), sharex=True, sharey=True)
        fig2.suptitle(
            f"Original CIL: {FUEL} | {MODEL_NAME} | Year={YEAR_PLOT}", y=0.98
        )

        for r, lg in enumerate([3, 2, 1]):
            for c, tr in enumerate([3, 2, 1]):
                ax     = axes2[r, c]
                subInc = inc_means[lg]
                subCDD = cdd_means[tr]
                subHDD = hdd_means[tr]

                y = yhat_interacted(
                    coef_orig, FUEL, subInc, subCDD, subHDD,
                    ibar, YEAR_PLOT, suffix=suffix_orig
                )
                ax.plot(T, y, linewidth=2, color="blue", alpha=0.8)
                ax.axhline(0, color="black", linewidth=0.5, linestyle="--")
                ax.axvline(20, color="gray",  linewidth=0.5, linestyle=":")
                ax.set_title(f"Inc={subInc:.2f}, CDD={subCDD:.0f}", fontsize=9)
                ax.grid(True, alpha=0.3)

                if c == 0:
                    ax.set_ylabel("ΔE", fontsize=10)
                if r == 2:
                    ax.set_xlabel("Temperature (°C)", fontsize=10)

        plt.tight_layout(rect=[0, 0, 1, 0.96])

    else:
        fig2, ax2 = plt.subplots(figsize=(10, 6))
        ax2.set_title(f"Original CIL: {FUEL} | Global model", fontsize=14)
        y = yhat_global(coef_orig, FUEL, suffix=suffix_orig)
        ax2.plot(T, y, linewidth=2, color="blue", alpha=0.8, label="Original GMFD")
        ax2.axhline(0, color="black", linewidth=0.5, linestyle="--")
        ax2.axvline(20, color="gray",  linewidth=0.5, linestyle=":")
        ax2.set_xlabel("Temperature (°C)", fontsize=12)
        ax2.set_ylabel("ΔE", fontsize=12)
        ax2.grid(True, alpha=0.3)
        ax2.legend(frameon=False)
        plt.tight_layout()

    out2 = (OUTDIR / f"ORIGINAL_{FUEL}_global.pdf"
            if MODEL_TYPE == "global" else
            OUTDIR / f"ORIGINAL_{FUEL}_{MODEL_NAME}.pdf")
    fig2.savefig(out2)
    print(f"\nSaved: {out2.name}")

else:
    print(f"\nWARNING: {orig_coef_path} not found — skipping original plot")

# ============================================================
# PLOT 3 — COMPARISON: original vs new GMFD
# ============================================================

# Only produced if both GMFD new coefficients and original coefficients
# were successfully loaded. Compares like-for-like (GMFD vs GMFD) to isolate
# differences due to pipeline changes rather than product differences.
if "GMFD" in coefs_new and coef_orig is not None:
    print("\n" + "=" * 60)
    print("PLOTTING COMPARISON (Original vs New GMFD)")
    print("=" * 60)

    if needs_climate_income:
        fig3, axes3 = plt.subplots(3, 3, figsize=(12, 9), sharex=True, sharey=True)
        fig3.suptitle(
            f"Comparison: Original (blue) vs New GMFD (red) | {FUEL} | {MODEL_NAME}",
            y=0.98
        )

        for r, lg in enumerate([3, 2, 1]):
            for c, tr in enumerate([3, 2, 1]):
                ax     = axes3[r, c]
                subInc = inc_means[lg]
                subCDD = cdd_means[tr]
                subHDD = hdd_means[tr]

                y_orig = yhat_interacted(
                    coef_orig, FUEL, subInc, subCDD, subHDD,
                    ibar, YEAR_PLOT, suffix=suffix_orig
                )
                y_new = yhat_interacted(
                    coefs_new["GMFD"], FUEL, subInc, subCDD, subHDD,
                    ibar, YEAR_PLOT, suffix=""
                )

                ax.plot(T, y_orig, linewidth=2, color="blue", alpha=0.7, label="Original")
                ax.plot(T, y_new,  linewidth=2, color="red",  alpha=0.7, label="New")
                ax.axhline(0, color="black", linewidth=0.5, linestyle="--")
                ax.axvline(20, color="gray",  linewidth=0.5, linestyle=":")
                ax.set_title(f"Inc={subInc:.2f}, CDD={subCDD:.0f}", fontsize=9)
                ax.grid(True, alpha=0.3)

                if c == 0:
                    ax.set_ylabel("ΔE", fontsize=10)
                if r == 2:
                    ax.set_xlabel("Temperature (°C)", fontsize=10)

        handles, labels = axes3[0, 0].get_legend_handles_labels()
        fig3.legend(handles, labels, loc="center right", frameon=False)
        plt.tight_layout(rect=[0, 0, 0.92, 0.96])

    else:
        fig3, ax3 = plt.subplots(figsize=(10, 6))
        ax3.set_title(
            f"Comparison: Original vs New GMFD | {FUEL} | Global", fontsize=14
        )
        y_orig = yhat_global(coef_orig,        FUEL, suffix=suffix_orig)
        y_new  = yhat_global(coefs_new["GMFD"], FUEL, suffix="")
        ax3.plot(T, y_orig, linewidth=2, color="blue", alpha=0.7, label="Original")
        ax3.plot(T, y_new,  linewidth=2, color="red",  alpha=0.7, label="New")
        ax3.axhline(0, color="black", linewidth=0.5, linestyle="--")
        ax3.axvline(20, color="gray",  linewidth=0.5, linestyle=":")
        ax3.set_xlabel("Temperature (°C)", fontsize=12)
        ax3.set_ylabel("ΔE", fontsize=12)
        ax3.grid(True, alpha=0.3)
        ax3.legend(frameon=False)
        plt.tight_layout()

    out3 = (OUTDIR / f"COMPARISON_GMFD_{FUEL}_global.pdf"
            if MODEL_TYPE == "global" else
            OUTDIR / f"COMPARISON_GMFD_{FUEL}_{MODEL_NAME}.pdf")
    fig3.savefig(out3)
    print(f"\nSaved: {out3.name}")

print("\n" + "=" * 60)
print("ALL DONE")
print("=" * 60)
print(f"\nFigures saved to: {OUTDIR}")

plt.show()
```
