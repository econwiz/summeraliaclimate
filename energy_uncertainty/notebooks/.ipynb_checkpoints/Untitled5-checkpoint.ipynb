```python
import numpy as np
import pandas as pd

# --------------------------
# 0) Inputs you already have
# --------------------------
# coef: pandas Series indexed by parm strings
# cell covariates (from your printed tables):
# subCDD, subHDD, subInc = ...
# ibar / anchor (submax) already computed from FD panel:
# submax = 9.087010110053141

# --------------------------
# 1) Helpers
# --------------------------
def get_beta(coef, name, default=0.0):
    """Safe coefficient getter."""
    return float(coef[name]) if name in coef.index else float(default)

def dc1_value(subInc, submax):
    """Remember: negative spline, clipped at 0 above the knot."""
    return min(subInc - submax, 0.0)

def income_group_from_dc1(dc1_val):
    """I1 below knot (dc1<0), I2 above knot (dc1==0)."""
    return 1 if dc1_val < 0 else 2

def temp_poly_terms(T):
    """Return [T, T^2] for your current spec (temp1,temp2)."""
    return T, T**2

# --------------------------
# 2) Choose ONE cell (same as your debug cell)
# --------------------------
# Example: income tercile lg=3, climate tercile tr=1
subInc = 9.997271537780762
subCDD = 2412.935302734375
subHDD = 156.36099243164062

submax = 9.087010110053141  # your printed anchor
dc1_val = dc1_value(subInc, submax)
IG = income_group_from_dc1(dc1_val)

print("CELL:")
print("  subInc =", subInc)
print("  subCDD =", subCDD)
print("  subHDD =", subHDD)
print("  submax(anchor) =", submax)
print("  dc1_val =", dc1_val)
print("  IG =", IG)

# --------------------------
# 3) Build y(T) for ONE pg and one fuel flag (your pg/indp logic)
# --------------------------
# You printed PG=1 and indf1; keep that for now.
PG = 1
F = 1  # indf1

# Temperature grid like paper figs
Tgrid = np.linspace(-15, 45, 301)

# Construct the *exact* coefficient names we will use
# (this matches your coef index style)
def nm(indp, indf, var):
    return f"c.indp{indp}#c.indf{indf}#c.{var}"

# In your earlier prints you were using indp1 for PG=1. Keep that.
indp = PG
indf = F

# Core betas (temp)
b_t1 = get_beta(coef, nm(indp, indf, "FD_temp1"))
b_t2 = get_beta(coef, nm(indp, indf, "FD_temp2"))

# TINV betas
b_cdd1 = get_beta(coef, nm(indp, indf, "FD_cdd20_TINVtemp1"))
b_cdd2 = get_beta(coef, nm(indp, indf, "FD_cdd20_TINVtemp2"))
b_hdd1 = get_beta(coef, nm(indp, indf, "FD_hdd20_TINVtemp1"))
b_hdd2 = get_beta(coef, nm(indp, indf, "FD_hdd20_TINVtemp2"))

# Income-spline betas (these should drop out when dc1_val==0)
b_dc1_t1 = get_beta(coef, nm(indp, indf, f"FD_dc1_lgdppc_MA15I{IG}temp1"))
b_dc1_t2 = get_beta(coef, nm(indp, indf, f"FD_dc1_lgdppc_MA15I{IG}temp2"))

print("\nPARAM CHECK:")
print("  b_t1, b_t2 =", b_t1, b_t2)
print("  b_cdd1, b_cdd2 =", b_cdd1, b_cdd2)
print("  b_hdd1, b_hdd2 =", b_hdd1, b_hdd2)
print("  b_dc1_t1, b_dc1_t2 =", b_dc1_t1, b_dc1_t2)

# Compute y(T) and also component contributions for debugging
y = np.zeros_like(Tgrid, dtype=float)

contrib_temp = np.zeros_like(y)
contrib_tinv = np.zeros_like(y)
contrib_dc1  = np.zeros_like(y)

for i, T in enumerate(Tgrid):
    t1, t2 = temp_poly_terms(T)

    # baseline temp part
    temp_part = b_t1 * t1 + b_t2 * t2

    # TINV interactions: multiply by cell-level CDD/HDD scalars
    tinv_part = (b_cdd1 * subCDD * t1 +
                 b_cdd2 * subCDD * t2 +
                 b_hdd1 * subHDD * t1 +
                 b_hdd2 * subHDD * t2)

    # income spline interactions (should be zero if dc1_val==0)
    dc1_part = dc1_val * (b_dc1_t1 * t1 + b_dc1_t2 * t2)

    contrib_temp[i] = temp_part
    contrib_tinv[i] = tinv_part
    contrib_dc1[i]  = dc1_part

    y[i] = temp_part + tinv_part + dc1_part

print("\nOUTPUT SANITY:")
print("  y min/max:", float(y.min()), float(y.max()))
print("  dc1 contrib min/max:", float(contrib_dc1.min()), float(contrib_dc1.max()))
print("  (dc1 should be exactly 0 everywhere if dc1_val==0)")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

COEF_PATH = Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty/data/regression/sters/FD_FGLS_global_GMFD_coeff.csv")

T = np.linspace(-15, 45, 600)

def read_coeff(path: Path) -> pd.Series:
    df = pd.read_csv(path)
    return pd.Series(df["beta"].values, index=df["parm"].astype(str).values)

def basis_from_builder(T):
    """
    Reproduce the *same functional forms* as your Python builder:
    - tempk = T^k
    - polyAbovek = (T^k - 20^k) if T>=20 else 0
    - polyBelowk = (20^k - T^k) if T<20 else 0
    """
    T = np.asarray(T)
    out = {}
    for k in [1,2,3,4]:
        Tk = T**k
        out[f"temp{k}"] = Tk
        out[f"polyAbove{k}"] = np.where(T >= 20, Tk - 20**k, 0.0)
        out[f"polyBelow{k}"] = np.where(T <  20, 20**k - Tk, 0.0)
    return out

def normalize_min_window(y, lo=10.0, hi=25.0):
    mask = (T >= lo) & (T <= hi)
    return y - np.min(y[mask])

coef = read_coeff(COEF_PATH)

# --- choose a block to debug ---
pg = 1        # indp1 or indp2 (whatever your plot dimension is)
ig = 2        # income-group side in the interacted spec (I1 vs I2)
indf = 1      # you're using indf1 in your printed parm names

B = basis_from_builder(T)

# Pull the exact coefficients that correspond to EACH basis component.
# Start with just temp1,temp2 and polyAbove/polyBelow terms.
def getb(term):
    name = f"c.indp{pg}#c.indf{indf}#c.FD_{term}"
    return float(coef.get(name, 0.0))

b_temp1 = getb("temp1")
b_temp2 = getb("temp2")

b_abv1  = getb("polyAbove1")
b_abv2  = getb("polyAbove2")
b_blw1  = getb("polyBelow1")
b_blw2  = getb("polyBelow2")

# income-spline interacted temp terms (if you want them later, add once dc1 is right)
# for now, turn them OFF until we trust shape
b_dc1_t1 = 0.0
b_dc1_t2 = 0.0

y_raw = (
    b_temp1 * B["temp1"] +
    b_temp2 * B["temp2"] +
    b_abv1  * B["polyAbove1"] +
    b_abv2  * B["polyAbove2"] +
    b_blw1  * B["polyBelow1"] +
    b_blw2  * B["polyBelow2"]
)

y = normalize_min_window(y_raw)

plt.figure(figsize=(7,4))
plt.plot(T, y, label="GMFD (pg block only)")
plt.axhline(0, linewidth=1)
plt.axvline(20, linewidth=1)
plt.xlim(-15, 45)
plt.title("Debug response using temp + polyAbove/polyBelow basis (normalized)")
plt.xlabel("T (°C)")
plt.ylabel("relative index (min over [10,25])")
plt.legend()
plt.tight_layout()
plt.show()

print("coeff debug:")
print("temp1,temp2:", b_temp1, b_temp2)
print("abv1,abv2:", b_abv1, b_abv2)
print("blw1,blw2:", b_blw1, b_blw2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# -------------------------
# Inputs you already have
# -------------------------
# cell_df: one row per (tgpid, tpid) with avgInc_tgpid, avgCDD_tpid, avgHDD_tpid
# Example: build from your printed tables (or from the "b_temp" df you had)
#
# Required columns: ["tgpid","tpid","avgInc_tgpid","avgCDD_tpid","avgHDD_tpid"]
#
# If you don't have it as a dataframe, make it like:
# cell_df = (
#   df[["tgpid","tpid","avgInc_tgpid","avgCDD_tpid","avgHDD_tpid"]]
#   .drop_duplicates()
#   .sort_values(["tgpid","tpid"])
# )

# -------------------------
# File paths
# -------------------------
COEF_PATHS = {
    "GMFD":   Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty/data/regression/sters/FD_FGLS_global_GMFD_coeff.csv"),
    "ERA5":   Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty/data/regression/sters/FD_FGLS_global_ERA5_coeff.csv"),
    "JRA_3Q": Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty/data/regression/sters/FD_FGLS_global_JRA_3Q_coeff.csv"),
    "MERRA2": Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty/data/regression/sters/FD_FGLS_global_MERRA2_coeff.csv"),
}

# Panel used only to compute the spline knot (ibar) the same way Stata did
FD_PANEL_PATH = Path("/user/ab5405/summeraliaclimate/code/energy_uncertainty/data/regression/GMFD_TINV_clim_regsort.dta")
b
FUEL = "electricity"  # set this once

# -------------------------
# Temperature grid
# -------------------------
TMIN, TMAX = -5.0, 35.0
T = np.linspace(TMIN, TMAX, 600)

def normalize_min_over_range(y, T, lo=10.0, hi=25.0):
    mask = (T >= lo) & (T <= hi)
    return y - np.min(y[mask])

# -------------------------
# Read coeffs
# -------------------------
def read_coeff(path: Path) -> pd.Series:
    df = pd.read_csv(path)
    # expected columns: parm, beta
    return pd.Series(df["beta"].to_numpy(), index=df["parm"].astype(str).to_numpy())

# -------------------------
# Knot (ibar) from panel: max lgdppc among largegpid==1 & product==fuel
# (matches your Stata logic for dc1_lgdppc_MA15)
# -------------------------
def compute_ibar_from_panel(panel_path: Path, fuel: str) -> float:
    import pyreadstat
    df, meta = pyreadstat.read_dta(str(panel_path), usecols=["product","largegpid","lgdppc_MA15"])
    sub = df[(df["largegpid"] == 1) & (df["product"] == fuel)]
    return float(sub["lgdppc_MA15"].max())

# -------------------------
# Basis functions up to degree K (autodetect from available coeffs)
# IMPORTANT: these are the *level* basis evaluated at T;
# we’re constructing the “implied response” index curve the same way we did before.
# -------------------------
def make_temp_basis(T, K):
    B = {}
    for k in range(1, K+1):
        B[f"temp{k}"] = T**k

    # polyAbove/polyBelow as in your builder (offset at 20)
    for k in range(1, K+1):
        B[f"polyAbove{k}"] = np.where(T >= 20.0, T**k - 20.0**k, 0.0)
        B[f"polyBelow{k}"] = np.where(T < 20.0, 20.0**k - T**k, 0.0)

    return B

def detect_max_degree(coef: pd.Series, pg: int) -> int:
    # look for FD_temp{k} terms
    ks = []
    for k in range(1, 5):  # your pipeline often has up to 4
        name = f"c.indp{pg}#c.indf1#c.FD_temp{k}"
        if name in coef.index:
            ks.append(k)
    return max(ks) if ks else 2  # safe fallback

# -------------------------
# Cell covariates
# -------------------------
def get_cell_covariates(cell_df: pd.DataFrame, tgpid: int, tpid: int):
    row = cell_df[(cell_df["tgpid"] == tgpid) & (cell_df["tpid"] == tpid)]
    if row.shape[0] != 1:
        raise ValueError(f"Expected exactly 1 row for tgpid={tgpid}, tpid={tpid}, got {row.shape[0]}")
    r = row.iloc[0]
    inc = float(r["avgInc_tgpid"])
    cdd = float(r["avgCDD_tpid"])
    hdd = float(r["avgHDD_tpid"])
    return cdd, hdd, inc

# -------------------------
# Income spline regime & dc1 value
# -------------------------
def income_spline(inc: float, ibar: float):
    # Stata: dc1 = lgdppc - max(...)  -> <=0 by construction for most obs
    dc1 = min(inc - ibar, 0.0)
    ig = 1 if inc <= ibar else 2
    return dc1, ig

# -------------------------
# Core: implied response curve for one product, one (tgpid,tpid), one pg block
# -------------------------
def implied_curve_cell(
    coef: pd.Series,
    *,
    pg: int,
    cdd: float,
    hdd: float,
    inc: float,
    ibar: float,
    include_tinv: bool = True,
    include_dc1: bool = True,
    normalize: bool = True,
):
    K = detect_max_degree(coef, pg=pg)
    B = make_temp_basis(T, K)

    # baseline temp terms: FD_temp{k}
    y = np.zeros_like(T, dtype=float)
    for k in range(1, K+1):
        key = f"c.indp{pg}#c.indf1#c.FD_temp{k}"
        if key in coef.index:
            y += float(coef[key]) * B[f"temp{k}"]

    # polyAbove/polyBelow baseline (if present in your regression)
    for k in range(1, K+1):
        keyA = f"c.indp{pg}#c.indf1#c.FD_polyAbove{k}"
        keyB = f"c.indp{pg}#c.indf1#c.FD_polyBelow{k}"
        if keyA in coef.index:
            y += float(coef[keyA]) * B[f"polyAbove{k}"]
        if keyB in coef.index:
            y += float(coef[keyB]) * B[f"polyBelow{k}"]

    # TINV interactions: CDD * polyAbove and HDD * polyBelow
    # (your exported coeff names look like FD_cdd20_TINVtemp1 etc.
    #  and in your debug you were mapping those onto polyAbove1/2 & polyBelow1/2)
    if include_tinv:
        for k in range(1, K+1):
            kc = f"c.indp{pg}#c.indf1#c.FD_cdd20_TINVtemp{k}"
            kh = f"c.indp{pg}#c.indf1#c.FD_hdd20_TINVtemp{k}"
            if kc in coef.index:
                y += float(coef[kc]) * cdd * B[f"polyAbove{k}"]
            if kh in coef.index:
                y += float(coef[kh]) * hdd * B[f"polyBelow{k}"]

    # Income spline: dc1 * temp terms (your names: FD_dc1_lgdppc_MA15I{ig}temp{k})
    if include_dc1:
        dc1, ig = income_spline(inc, ibar)
        if dc1 != 0.0:
            for k in range(1, K+1):
                kd = f"c.indp{pg}#c.indf1#c.FD_dc1_lgdppc_MA15I{ig}temp{k}"
                if kd in coef.index:
                    y += float(coef[kd]) * dc1 * B[f"temp{k}"]

    # normalize to match “index” style (min over 10–25C)
    if normalize:
        y = normalize_min_over_range(y, T, lo=10.0, hi=25.0)

    return y

# -------------------------
# 3×3 plotter: overlays products inside each cell
# -------------------------
def plot_3x3(cell_df: pd.DataFrame, *, pg: int = 2, include_tinv=True, include_dc1=True):
    # compute ibar once
    ibar = compute_ibar_from_panel(FD_PANEL_PATH, FUEL)
    print("ibar =", ibar)

    # load coeffs once
    COEFS = {prod: read_coeff(path) for prod, path in COEF_PATHS.items()}

    fig, axes = plt.subplots(3, 3, figsize=(12, 9), sharex=True, sharey=True)

    for r, tgpid in enumerate([1, 2, 3]):      # rows = income terciles
        for c, tpid in enumerate([1, 2, 3]):   # cols = climate terciles
            ax = axes[r, c]

            cdd, hdd, inc = get_cell_covariates(cell_df, tgpid=tgpid, tpid=tpid)
            ax.set_title(f"Income tgpid={tgpid}, Climate tpid={tpid}\nInc={inc:.2f}, CDD={cdd:.0f}, HDD={hdd:.0f}", fontsize=9)

            for prod, coef in COEFS.items():
                y = implied_curve_cell(
                    coef,
                    pg=pg,
                    cdd=cdd, hdd=hdd, inc=inc, ibar=ibar,
                    include_tinv=include_tinv,
                    include_dc1=include_dc1,
                    normalize=True,
                )
                ax.plot(T, y, linewidth=2, label=prod)

            ax.axvline(20, linewidth=1)
            ax.axhline(0, linewidth=1)
            ax.grid(alpha=0.2)

            if r == 2:
                ax.set_xlabel("Temperature (°C)")
            if c == 0:
                ax.set_ylabel("Index (normalized)")

    # legend once
    handles, labels = axes[0,0].get_legend_handles_labels()
    fig.legend(handles, labels, loc="center right", frameon=False)
    fig.suptitle(f"3×3 Energy–Temperature response (pg={pg}, fuel={FUEL})\n"
                 f"TINV={'ON' if include_tinv else 'OFF'} | dc1={'ON' if include_dc1 else 'OFF'} | normalized min[10,25]",
                 y=0.98)
    plt.tight_layout(rect=[0,0,0.88,0.95])
    plt.show()

# -------------------------
# Run it:
# -------------------------
# plot_3x3(cell_df, pg=2, include_tinv=True, include_dc1=False)  # start safer
# then:
# plot_3x3(cell_df, pg=2, include_tinv=True, include_dc1=True)
```
